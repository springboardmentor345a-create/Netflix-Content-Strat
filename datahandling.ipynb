```python
import pandas as pd
df = pd.read_csv("netflix_titles .csv")
# 1. Convert the string to a Datetime object first
# We use errors='coerce' to turn any messy data into 'NaT' (Not a Time)
df['date_added'] = pd.to_datetime(df['date_added'].str.strip(), errors='coerce')

# 2. Extract the year and fill any missing values with 0 so we can convert to integer
# (Since you already dropped nulls in the previous step, this is just a safety measure)
df['date_added'] = df['date_added'].dt.year.fillna(0).astype('int64')

# 3. Verify the change
print(df[['title', 'date_added']].info())
df[['title', 'date_added']].head()
df.info()
```

```python
# 1. Check for Null Values
print("--- Null Value Check ---")
print(df.isnull().sum())

# 2. Check for Duplicate Rows
# (Netflix data sometimes has duplicates if a title was re-added)
duplicates = df.duplicated().sum()
print(f"\n--- Duplicate Check ---")
print(f"Total duplicate rows found: {duplicates}")

# 3. Final Type Verification
print(f"\n--- Data Type Check ---")
print(df[['date_added']].dtypes)

# 4. If duplicates exist, run this to remove them:
if duplicates > 0:
    df.drop_duplicates(inplace=True)
    print("\nAction: Duplicates removed.")
```

```python
# 1. Permanent Replacement for categorical columns
# We assign the result back to the columns to ensure they 'stick'
df['director'] = df['director'].fillna('notavailable')
df['cast'] = df['cast'].fillna('notavailable')
df['country'] = df['country'].fillna('notavailable')

# 2. Permanent Drop for the small null counts (Rating and Duration)
# 'inplace=True' makes the change directly to the dataframe
df.dropna(subset=['rating', 'duration'], inplace=True)

# 3. FINAL VERIFICATION
print("--- Final Null Value Count ---")
print(df.isnull().sum())
```

```python
# 1. Check for Null Values
print("--- Null Value Check ---")
print(df.isnull().sum())

# 2. Check for Duplicate Rows
# (Netflix data sometimes has duplicates if a title was re-added)
duplicates = df.duplicated().sum()
print(f"\n--- Duplicate Check ---")
print(f"Total duplicate rows found: {duplicates}")

# 3. Final Type Verification
print(f"\n--- Data Type Check ---")
print(df[['date_added']].dtypes)

# 4. If duplicates exist, run this to remove them:
if duplicates > 0:
    df.drop_duplicates(inplace=True)
    print("\nAction: Duplicates removed.")
```

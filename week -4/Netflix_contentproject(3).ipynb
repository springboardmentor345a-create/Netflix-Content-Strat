```python
null_values = df.isnull().sum()
print("Null Values in Each Column:\n")
print(null_values)
```

```python
df['director'] = df['director'].fillna("Unknown")
df['cast'] = df['cast'].fillna("Not Available")
df['country'] = df['country'].fillna("Not Available")
df['rating'] = df['rating'].fillna("not Rated")
```

```python
null_values = df.isnull().sum()
print(null_values)
```

```python
#Drop Very Small Null Rows
df=df.dropna(subset=['duration'])
df = df.dropna(subset=['date_added'])
```

```python
null_values = df.isnull().sum()
print(null_values)
```

```python
duplicate_count = df.duplicated().sum()

print("Number of Duplicate Rows:", duplicate_count)
```

```python
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')
df['release_year'] = pd.to_datetime(df['release_year'], format='%Y')
```

```python
print("After Conversion:")
print("release_year datatype:", df['release_year'].dtype)
print("date_added datatype:", df['date_added'].dtype)
```

```python
print(df.info())
```

```python
df['date_added'] = df['date_added'].fillna("Unknown")
df['date_added'].isnull().sum()
```

```python
df.columns
```

```python
#Case Inconsistency
print(df['type'].unique())
print(df['rating'].unique())
```

```python
df['rating'] = df['rating'].replace({'UR': 'NR'})
print(df['rating'].unique())
```

```python
#Check Numeric Column Consistency
print("Min Year:", df['release_year'].min())
print("Max Year:", df['release_year'].max())
```

```python
#Check Logical Category Errors
expected_types = ['Movie', 'TV Show']
print("Unexpected Type Values:",
      set(df['type']) - set(expected_types))
```

```python
#Categorical Columns Identification
categorical_columns = df.select_dtypes(include='object').columns
print("Categorical Columns:", categorical_columns)
```

```python
df['duration_int'] = df['duration'].str.extract('(\d+)')
df['duration_int'] = df['duration_int'].astype(float)
```

```python
df['content_length_category'] = df['duration_int'].apply(
    lambda x: 'Short' if x < 60 else ('Medium' if x < 120 else 'Long')
)
```

```python
df['primary_genre'] = df['listed_in'].apply(lambda x: x.split(',')[0])
```

```python
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')
```

```python
df['year_added'] = df['date_added'].dt.year
df['month_added'] = df['date_added'].dt.month
```

```python
df['country'] = df['country'].apply(lambda x: x.split(',')[0])
```

```python
df['primary_genre'] = df['listed_in'].apply(lambda x: x.split(',')[0])
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(6,6))

df['type'].value_counts().plot.pie(autopct='%1.1f%%')

plt.title("Movies vs TV Shows on Netflix")
plt.ylabel("")
plt.show()
```

```python
top_genres = df['primary_genre'].value_counts().head(10)

plt.figure(figsize=(10,5))

sns.barplot(x=top_genres.values, y=top_genres.index)

plt.title("Top 10 Genres on Netflix")
plt.xlabel("Number of Titles")
plt.ylabel("Genre")
plt.show()
```

```python
top_countries = df['country'].value_counts().head(10)

plt.figure(figsize=(10,5))

sns.barplot(x=top_countries.values, y=top_countries.index)

plt.title("Top Countries Producing Netflix Content")
plt.xlabel("Number of Titles")
plt.ylabel("Country")
plt.show()
```

```python
plt.figure(figsize=(10,5))

sns.countplot(data=df, x='rating')

plt.title("Content Rating Distribution")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10,5))

sns.countplot(data=df, x='year_added')

plt.title("Content Added to Netflix Each Year")
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.countplot(data=df, x='content_length_category')

plt.title("Content Length Distribution")
plt.xlabel("Content Length")
plt.ylabel("Number of Titles")

plt.show()
```

```python
genre_year = df.groupby(['year_added','primary_genre']).size().reset_index(name='count')

top_genres = df['primary_genre'].value_counts().head(5).index

genre_year = genre_year[genre_year['primary_genre'].isin(top_genres)]

plt.figure(figsize=(10,6))

sns.lineplot(data=genre_year, x='year_added', y='count', hue='primary_genre')

plt.title("Top Genre Trends Over Years")

plt.show()
```

```python
country_type = pd.crosstab(df['country'], df['type']).head(10)

country_type.plot(kind='bar', figsize=(10,6))

plt.title("Country vs Content Type")

plt.show()
```

```python
type_year = df.groupby(['year_added','type']).size().reset_index(name='count')

plt.figure(figsize=(10,6))

sns.lineplot(data=type_year, x='year_added', y='count', hue='type')

plt.title("Movies vs TV Shows Added Over Time")

plt.show()
```

Create Derived Feature: Content Length Category

```python
df['duration_num'] = df['duration'].str.extract('(\d+)').astype(float)
```

```python
def length_category(x):
    if x < 60:
        return "Short"
    elif x < 120:
        return "Medium"
    else:
        return "Long"

df['content_length_category'] = df['duration_num'].apply(length_category)
```

Clustering Netflix Titles (Genre, Duration, Rating)

```python
from sklearn.preprocessing import LabelEncoder

# Encode rating
le_rating = LabelEncoder()
df['rating_encoded'] = le_rating.fit_transform(df['rating'])

# Encode genre (listed_in)
le_genre = LabelEncoder()
df['genre_encoded'] = le_genre.fit_transform(df['listed_in'])

# Extract numeric duration
df['duration_num'] = df['duration'].str.extract(r'(\d+)').astype(float)
```

```python
X_cluster = df[['rating_encoded','genre_encoded','duration_num']].dropna()
```

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=3, random_state=42)

df.loc[X_cluster.index, 'cluster'] = kmeans.fit_predict(X_cluster)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,6))

scatter = plt.scatter(
    X_cluster['duration_num'],
    X_cluster['rating_encoded'],
    c=df.loc[X_cluster.index,'cluster'],
    cmap='viridis',
    alpha=0.7
)

plt.xlabel("Duration (Minutes / Seasons)")
plt.ylabel("Encoded Rating")
plt.title("Netflix Content Clusters based on Duration and Rating")

plt.colorbar(scatter, label="Cluster Group")

plt.grid(True)

plt.show()
```

Classification (Movie vs TV Show)

```python
from sklearn.preprocessing import LabelEncoder

# Encode rating
le_rating = LabelEncoder()
df['rating_encoded'] = le_rating.fit_transform(df['rating'])

# Encode genre
le_genre = LabelEncoder()
df['genre_encoded'] = le_genre.fit_transform(df['listed_in'])

# Encode type (target)
le_type = LabelEncoder()
df['type_encoded'] = le_type.fit_transform(df['type'])
```

```python
df['duration_num'] = df['duration'].str.extract(r'(\d+)').astype(float)
```

```python
features = df[['rating_encoded','genre_encoded','duration_num']].dropna()

target = df.loc[features.index,'type_encoded']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    features, target, test_size=0.2, random_state=42
)
```

```python
from sklearn.tree import DecisionTreeClassifier

model = DecisionTreeClassifier()

model.fit(X_train, y_train)
```

```python
accuracy = model.score(X_test, y_test)

print("Model Accuracy:", accuracy)
```

Analyze Key Drivers for Content Availability

Country-Level Content Availability

```python
import pandas as pd

importance = pd.Series(model.feature_importances_,
                       index=features.columns)

importance.sort_values().plot(kind='barh')

plt.title("Feature Importance")
plt.xlabel("Importance Score")
plt.show()
```

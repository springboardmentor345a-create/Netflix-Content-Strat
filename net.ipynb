```python
import pandas as pd
df=pd.read_csv(r"C:\Users\iqbal\Downloads\netflix_titles.csv\netflix_titles.csv")
df.head()
```

```python
print(f"Dataset Shape: {df.shape}\n")
df.describe()
df.info()
```

```python
print(df.isnull().sum())
df['director'] = df['director'].fillna('Unknown')
df['cast'] = df['cast'].fillna('Unknown')
df['country'] = df['country'].fillna('Unknown')
df = df.dropna(subset=['date_added', 'rating', 'duration'])
df['date_added'] = pd.to_datetime(df['date_added'].str.strip())
print(df.isnull().sum())
```

```python
df['year_added'] = df['date_added'].dt.year
df['month_added'] = df['date_added'].dt.month_name()
movies_df = df[df['type'] == 'Movie'].copy()
movies_df['duration_minutes'] = movies_df['duration'].str.replace(' min', '').astype(int)
movies_df[['title', 'duration', 'duration_minutes']].head()
```

```python
columns_to_fill = ['director', 'cast', 'country']
for col in columns_to_fill:
    df[col] = df[col].fillna('Unknown')
df = df.dropna(subset=['date_added', 'rating', 'duration'])

print(df.isnull().sum())
```

```python
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')

df['release_year'] = pd.to_datetime(df['release_year'], format='%Y', errors='coerce')

print(df[['date_added', 'release_year']].dtypes)

df[['title', 'date_added', 'release_year']].head()
```

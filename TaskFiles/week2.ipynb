```python
import pandas as pd

# Load the dataset
df = pd.read_csv('netflix_titles.csv')

# 1. Fill heavily missing text columns with "Unknown"
# We must do this first so we don't accidentally drop thousands of rows
df['director'] = df['director'].fillna('Unknown')
df['cast'] = df['cast'].fillna('Unknown')
df['country'] = df['country'].fillna('Unknown')

# 2. Drop rows with missing values only in specific columns (Mentor's approach)
# The subset parameter ensures we only look at these specific columns for nulls
df.dropna(subset=['rating', 'duration', 'date_added'], inplace=True)

# 3. Verify the cleaning
print("--- Missing Values After Cleaning ---")
print(df.isnull().sum())

print(f"\nFinal Dataset Shape: {df.shape[0]} rows and {df.shape[1]} columns")
```

```python
df.shape
```

```python
# --- 3. Normalize Categorical Features ---

# Strip leading/trailing whitespaces and ensure consistent casing
# We use Title Case for countries and genres, and UPPERCASE for ratings
df['country'] = df['country'].str.strip().str.title()
df['listed_in'] = df['listed_in'].str.strip().str.title()
df['rating'] = df['rating'].str.strip().str.upper()

# Normalize spacing around commas for titles with multiple countries or genres
# This turns "United States,India" or "United States , India" into consistently formatted "United States, India"
df['country'] = df['country'].str.replace(r'\s*,\s*', ', ', regex=True)
df['listed_in'] = df['listed_in'].str.replace(r'\s*,\s*', ', ', regex=True)

# Verify the normalization (looking at a few unique values)
print("--- Normalized Sample Data ---")
print("Ratings:", df['rating'].unique()[:5])
print("Countries:", df['country'].unique()[:3])

# --- 4. Save the Final Cleaned Dataset ---
df.to_csv('netflix_cleaned.csv', index=False)
print("\nCleaned and normalized dataset saved successfully!")
```
